```python
import numpy as np
import pandas as pd
from faker import Faker
import random
fake = Faker('en_IN')

value = 10
add_list = ["delhi", "noida", "gurgaon"]
selected_add = random.choices(add_list, k=value)

add_prof = ["it", "business", "gov"]
selected_prof = random.choices(add_prof, k=value)

add_cont = ["india", "nepal", "siri lanka"]
selected_cont = random.choices(add_cont, k=value)

fake_data = {
    'id  ' : np.arange(1,value+1),
    'name' :  [fake.name() for _ in range(value)],
    'email' : [fake.email() for _ in range(value)],
    'phone' : [fake.phone_number() for _ in range(value)],
    'age': np.random.randint(20,30,size=value),
    'profession' : selected_prof,
    'salary': np.random.randint(30000,70000,size=value),
    'address' : selected_add,
    'country' : selected_cont,

}

df = pd.DataFrame(fake_data)
df
```

```python
import sqlite3

conx = sqlite3.connect('faker.db')
cur = conx.cursor()
```

```python
df.to_sql("fakertbl", conx, if_exists="replace", index=False)

conx.commit()
conx.close()
```

```python

conx = sqlite3.connect('faker.db')
cur = conx.cursor()

data = "SELECT * FROM fakertbl"
cur.execute(data)
d = cur.fetchall()
# print(d)
for i in d:
    print(i)

cur.close()
conx.commit()
```
